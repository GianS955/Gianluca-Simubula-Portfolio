# Taxi Driver

This project aims to explore and implement core Reinforcement Learning techniques within a discrete environment. 
The agent moves across the grid, avoiding obstacles, and completing a defined task involving pick-up and drop-off operations.

The environment consists of cells with different properties (free cells, obstacles, pick-up/drop-off points, destinations, etc.):

The cells divide into three types:
* **Roads**: most typical cell type. They allow the agent of transpassing them
* **Buildings**: they are considered as obstacles as the agent can not pass through them
* **Pick-up/drop-offs**: as road cells they can be freely passed on. Passangers to pick up can only be found on this cells and their destination is a different *drop-off* cell

Road and pick-up/drop-off cells can have obstacles too. 
They hinder the passage to the agent, and they can be of *N*. *S*, *W* or *E* type so that they block the passage of the corresponding direction.
One cell can have multiple obstacles.

The state of the agent is composed by:
* **Cell coordinates**: i-th row and j-th column of the cell where the agent is currently located
* **Passenger location**: n-th pick-up/drop-off cell where the passenger can be picked us on. Alternative &#8734; if the passenger is currently carried by the agent
* **Destination location**: m-th pick-up/drop-off cell where to drop-off the passenger

At each time step the agent can perform one of the following actions:
* **Drive**: move across the grid, following the *N*. *S*, *W* or *E* directions. This action holds reward equal to -1 if the next cell is either of *road* type of no obstacle hinder the movement, and -100 otherwise
* **Pick-up**: pick-up a passenger. This action can only be performed it the agent finds itself on a *pick-up/drop-off* cell. The recieved rewards is 20
* **Drop-off**: let the passenger on the grid. If the passenger hops of the agent on the desired destination, the latter recieve 40 and -10 otherwise.

The episode ends when the passenger is delivered on the desired destination.
